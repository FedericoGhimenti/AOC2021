```python
from statistics import mode

f = open('input14.txt')

lines = f.readlines()

polymer = lines[0].rstrip()

rule = []

for i in range(2,len(lines)):
    
    l = lines[i].rstrip()
    l = l.split(' -> ')
    rule.append(l[0])
    rule.append(l[1])
```

```python
def Polymerization(polymer, rules):
    
    new_polymer = str()
    
    polymer_length = len(polymer)
    
    for i in range(0, polymer_length - 1):
            
        couple = polymer[i] + polymer[i+1]
        
        insertion = rule[rule.index(couple) + 1]
        
        new_polymer += polymer[i] + insertion
        
    new_polymer += polymer[i+1]
        
    return new_polymer

def LetterCounter(polymer, letters):

    counts = [0] * len(letters)
        
    for c in polymer:

        counts[ letters.index(c) ] += 1
        
    return counts
  
    
def MostAndLeastCommon(counts, letters):

    M = max(counts)
    m = min(counts)
    
    most_common = letters[ counts.index(M) ]
    least_common = letters[ counts.index(m) ]
    print('The most common letter is ' + most_common + ' and appears ' + str(M) + ' times')
    print('The least common letter is ' + least_common + ' and appears ' + str(m) + ' times')
    print('The score is ' + str(M - m))
    
def Divide(polymer):
        
    new_poymer = []
    new_polymer = []
    
    length_polymer = len(polymer)
    
    for i in range(0,len(polymer)-1):
        
        new_polymer.append(polymer[i] + polymer[i+1])
        
    return new_polymer

def SeparatedPolymerization(polymer, rule, N_steps, N_substeps, counter):
    
    polymer = Divide(polymer)
    new_polymer = str()

    for segment in polymer:
        
        for i in range(0, N_substeps):
            
            segment = Polymerization(segment, rule)
            
        new_polymer += segment
        
    return new_polymer                 
```
