```python
import numpy as np

f = open('input8.txt')
data = []
lines = f.readlines()
for l in lines:
    i = l.index('|')
    l = l[i+2:]
    l = l.split(' ')
    l[-1] = l[-1].rstrip('\n')
    data.append(l)
```

```python
counter = 0
target = [2,3,4,7]
for l in data:
    for s in l:
        length = len(s)
        if length in target:
            counter += 1
print(counter)
```

```python
f = open('input8.txt')
data_output = []
data_input = []
lines = f.readlines()
for l in lines:
    
    i = l.index('|')
    
    out = l[i+2:]
    out = out.split(' ')
    out[-1] = out[-1].rstrip('\n')
    data_output.append(out)
    
    inp = l[:i-1]
    inp = inp.split(' ')
    inp.sort(key = len)
    data_input.append(inp)
    print(data_input)
```

```python
code_list = []
for l in data_input:
    code = ['a']*10
    code[1] = l[0]
    code[7] = l[1]
    code[4] = l[2]
    code[8] = l[9]
    
    for i in range(3,6):
        s = l[i]
        if all(c in s for c in code[1]) == True:
            code[3] = s
            
    for i in range(6,9):
        s = l[i]
        if all(c in s for c in code[3]) == True:
            code[9] = s
        elif all(c in s for c in code[1]) == True:
            code[0] = s
        else:
            code[6] = s
    
    for i in range(3,6):
        s = l[i]
        if len(list(set(s).difference(set(code[9])))) == 1 and s != code[3]:
            code[2] = s
        elif s != code[3]:
            code[5] = s
    print(code)
    code_list.append(code)   
```

```python
final_answer = 0
L = len(data_output)
for i in range(0,L):
    line_answer = str()
    code = code_list[i]
    l = data_output[i]
    
    for output in l:
        digit = 0
        for c in code:
            if set(c) == set(output):
                line_answer += str(digit)
                print(line_answer,digit,l,output,code)
            digit += 1
    final_answer += int(line_answer)
```
